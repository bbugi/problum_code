# 2738 - 행렬 덧셈

## 문제
* N*M크기의 두 행렬 A와 B가 주어졌을 때, 두 행렬을 더하는 프로그램을 작성하시오.  

## 입력
* 첫째 줄에 행렬의 크기 N 과 M이 주어진다. 둘째 줄부터 N개의 줄에 행렬 A의 원소 M개가 차례대로 주어진다.  
  이어서 N개의 줄에 행렬 B의 원소 M개가 차례대로 주어진다. N과 M은 100보다 작거나 같고, 행렬의 원소는 절댓값이 100보다 작거나 같은 정수이다.  
  
## 출력 
* 첫째 줄부터 N개의 줄에 행렬 A와 B를 더한 행렬을 출력한다. 행렬의 각 원소는 공백으로 구분한다.

## 풀이

````python
## 최종 제출 코드
# 행렬 만들기
a = []
b = []
n,m = map(int, input().split())
for _ in range(n):
    a.append(list(map(int, input().split(' '))))
for _ in range(n):
    b.append(list(map(int, input().split(' '))))
    
# 행렬 더하기
for i in range(n): # row
    for j in range(m): # col
        print(a[i][j] + b[i][j], end = ' ') # end = ' '으로 설정하여 한줄씩 출력되는 것을 띄어쓰기로 변경
    print() # 한 row가 출력되면 줄바꿈 수행
````

````python
## 런타임 에러 발생
# 행렬 만들기
n,m = map(int, input().split(' '))
a = []
b = []
for _ in range(n):
    a.append(list(map(int, input().split(' '))))
for _ in range(n):
    b.append(list(map(int, input().split(' '))))
# print(a)
# print(b)

# 행렬 더하기
for i in range(n):
    for j in range(m):
        a[i][j] += b[i][j]
# print(a)

# 배열 출력하기
for x, y, z in a:
    print(x, y, z)
````

# 1065 - 한수

## 문제
* 어떤 양의 정수 X의 각 자리가 등차수열을 이룬다면, 그 수를 한수라고 한다.  
  등차수열은 연속된 두 개의 수의 차이가 일정한 수열을 말한다.  
  N이 주어졌을 때, 1보다 크거나 같고, N보다 작거나 같은 한수의 개수를 출력하는 프로그램을 작성하시오. 
## 입력
* 첫째 줄에 1,000보다 작거나 같은 자연수 N이 주어진다.
## 출력
* 첫째 줄에 1보다 크거나 같고, N보다 작거나 같은 한수의 개수를 출력한다.

## 풀이

````python
n = int(input())
print(n)
````

````python
sum = 0
for i in range(1,n+1):
    # print(i)
    # 한자리 수 -> 무조건 등차수열
    # 두자리 수 -> 무조건 등차수열
    # 세자리 수 -> 등차수열 확인해야함
    if i < 10:
        sum += 1
    elif i < 100:
        sum += 1
    # else:
    #     a = list(map(int, str(i)))
print(sum)
````

````python
sum = 0
for i in range(1,n+1):
    if i >= 100:
        a = list(map(int, str(i)))
        if (a[2] - a[1]) == (a[1] - a[0]):
            sum +=1
print(sum)
        
````

````python
# 최종 정답 합치기
# 한수 정의 참조 https://www.acmicpc.net/board/view/25689

n = int(input())
sum = 0
for i in range(1,n+1):
    if i < 10:
        sum += 1
    elif i < 100:
        sum += 1
    else:
        a = list(map(int, str(i)))
        if (a[2] - a[1]) == (a[1] - a[0]):
            sum +=1
print(sum)
````
